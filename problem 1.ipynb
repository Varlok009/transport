__1.__ Метрики оценки развития транспортной инфраструктуры.

   __1.1__ Метрика количества использования ОТ

    - Необходимо на основании матриц корреспонденций сравнить количество поездок на ОТ до и после реализации мероприятий (как в рамках всего объема перевозок, так и для зон, которые затрагивает новый проект).
    - На основании изменения количества использования ОТ можно рассчитать дополнительную прибыль и при необходимости/возможности внести корректировки (например скорректировать маршруты наземного транспорта)
    
    - Для реализации расчетов достаточно матриц корреспонденций.

   __1.2__ Метрика изменения продолжительности передвижения от дома до работы

    - Для районов, расположенных вблизи с реализуемыми мероприятиями, необходимо собрать данные о проживающих: гео данные места проживания и места работы. Далее сравнить время в пути от дома до работы до реализации мероприятий и после. На основании среднего изменения времени в пути подготовить отличный прессрелиз о том, что для N жителей время до работы сократилось на X минут.
    
    - Для рассчетов потребуются геоданные пользователей от операторов сотовой связи, на их основании можно будет сделать предполижение о том, что гео данные пользователя в ночное время соответствует месту его проживания, а в дневное - месту работы.
    - Для рассчета времени в пути потенциально возможно использовать api яндекс карт.

   __1.3__ Метрика количества использования городских парковок

    - Для зон расположенных вблизи с реализуемыми мероприятиями необходимо подсчитать разницу в количестве использования парковок до и после реализации мероприятий. В зависимости от результатов можно будет планировать развитие парковочных пространств.
    
    - Для рассчетов необходимы данные о количестве использований городских парковок в прилегающих районах.
