__3.__ Сервис получения данных

Сервис будет состоять из 3 сущностей: **ftp** хранилище данных, **api** предобработки данных и их передачи клиенту, **клиентская** часть для простого доступа к данным и их визуализации.

  **1.** ftp сервер будет осуществлять структурированное хранение исходных данных

  **2.** api
    **2.1** В первую очередь необходимо реализовать выгрузку данных с ftp по расписанию, например с помощью cellery
    **2.2** Далее нужно предобработать данные: проверить формат данных и их целостность
    **2.3** Проверенные данные предлогается загружать в БД
    **2.4** Основное назначение: предоставление данных по запросу. Так например по запросу api/v1/peaks/34/last - будет передана информация для построения отчета о пиках в зоне 34 на основе последних данных (вчераший день).

  **3.** Клиентская часть предлагается на основе django приложение, которое позволит достаточно легко организовать аунтификацию пользователей и настроить для них доступ в зависимости от потребностей.

В личном кабинете будут стандартные кнопки для выгрузки отчетов, так после нажатия на кнопку "Вчерашние часы пик для города", на странице появится отчет с графиком (аналогичный отчету из задания 2.1), так же будет возможность выгрузить отчет в pdf.

Дополнительно можно отобразить карту районов города, при нажатии на районы такой карты будет всплывать краткая информация из отчета.
